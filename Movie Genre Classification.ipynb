```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
df['description'][0]
```

```python
split_data = df['description'][0].split()
```

```python
print(split_data)
```

```python
lower = []
for i in split_data:
    lower.append(i.lower())
print(lower)
```

```python
ps = PorterStemmer()
stemmed = []
for i in lower:
    stemmed.append(ps.stem(i))
result = ' '.join(stemmed)
print(result)
```

```python
def preprocess(text):
    stemmed = []
    for i in text.split():
        lower = i.lower()
        stemmed.append(ps.stem(lower))
    return ' '.join(stemmed)
    
```

```python
preprocess(df['description'][0])
```

```python
df['description'] = df['description'].apply(preprocess)
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import pandas as pd

def re_stop(text):
    try:
        nltk.data.find('corpora/stopwords')
        nltk.data.find('tokenizers/punkt')
    except LookupError:
        nltk.download('stopwords')
        nltk.download('punkt')

    words = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)


re_stop(df['description'][0])
df['description'] = df['description'].apply(re_stop)
df.head()
```

```python
df['title']= data['title']
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Assuming `df` is your DataFrame with the preprocessed `description`
vectorizer = TfidfVectorizer(max_features=5000)  # You can adjust max_features as needed
x = vectorizer.fit_transform(df['description'])
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df['genre'])
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x, y)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(x)

# Accuracy
accuracy = accuracy_score(y, y_pred)
print(f'Accuracy: {accuracy:.4f}')

# Classification report
print(classification_report(y, y_pred, target_names=label_encoder.classes_))
```
