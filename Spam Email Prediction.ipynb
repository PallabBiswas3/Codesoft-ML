```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('C:\\Users\\PALLAB\\Downloads\\Spam SMS\\spam.csv', encoding='ISO-8859-1')
df['v1'] = df['v1'].map({'spam':0 ,'ham':1})
data = df[['v1' , 'v2']]
data
```

# Removing StopWords

```python
import string
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))  # Removing punctution
    stop_words = set(stopwords.words('english'))                      # Removing stopwords
    text = ' '.join(word for word in text.split() if word not in stop_words)
    return text

df['v2'] = df['v2'].apply(preprocess_text)
```

#  Converting raw text documents into matrix of TF-IDF 

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000) 

x = tfidf.fit_transform(df['v2']).toarray()

y = df['v1']
```

```python

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Using ComplementNB Algorithm

```python
from sklearn.naive_bayes import ComplementNB 

com_model = ComplementNB ()

com_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

y_pred = com_model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')

print(classification_report(y_test, y_pred))
```

```python
new_email = ["Congratulations! You've won a $1000 gift card. Click here to claim your prize."]

new_email_processed = tfidf.transform([preprocess_text(new_email[0])]).toarray()
prediction = com_model.predict(new_email_processed)
print('Spam' if prediction[0] == 0 else 'Not Spam')
```
